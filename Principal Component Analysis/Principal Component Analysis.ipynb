# Principle Components Analysis of the NYTimes News Articles

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Import principle components

```python
pca = pd.read_csv("pca.csv", index_col=0)
pca.head(10)
```

### Split the principle components by years

```python
pca2011 = pca.loc[0:1878]
pca2012 = pca.loc[1879:3700]
pca2013 = pca.loc[3701:5499]
pca2014 = pca.loc[5500:7363]
pca2015 = pca.loc[7364:9156]
pca2016 = pca.loc[9157:10327]
pca2017 = pca.loc[10328:11162]
pca2018 = pca.loc[11163:]
```

## Comparisons of different years with the first and second PC

### PC Plot 2011, 2012, and 2013

```python
fig, ax = plt.subplots(figsize=(15,10))
ax.scatter(pca2011["0"], pca2011["1"], s = 10, c = "red", alpha = 0.5, linewidth = 0)
ax.scatter(pca2012["0"], pca2012["1"], s = 10, c = "blue", alpha = 0.5, linewidth = 0)
ax.scatter(pca2013["0"], pca2013["1"], s = 10, c = "green", alpha = 0.5, linewidth = 0)
ax.legend(["2011", "2012", "2013"])
ax.grid()
```

### PC Plot 2012, 2013, and 2014

```python
fig, ax = plt.subplots(figsize=(15,10))
ax.scatter(pca2012["0"],
           pca2012["1"],
           s = 10,
           c = "red",
           alpha = 0.5,
           linewidth = 0)
ax.scatter(pca2013["0"],
           pca2013["1"],
           s = 10,
           c = "blue",
           alpha = 0.5,
           linewidth = 0)
ax.scatter(pca2014["0"],
           pca2014["1"],
           s = 10,
           c = "green",
           alpha = 0.5,
           linewidth = 0)
ax.legend(["2012", "2013", "2014"])
ax.grid()
```

### PC Plot 2013, 2014, and 2015
